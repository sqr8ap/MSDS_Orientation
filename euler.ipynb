Problem 2

```python
# make fibo sequence
fibo = [1,2]
for i in range(1,10000):
    new_item = fibo[-1] + fibo[-2]
    fibo.append(new_item)
    if new_item > 4000000:
        break

# sum even terms
my_sum = 0
for item in fibo:
    if item % 2 == 0:
        my_sum += item

print(my_sum)
```

Convergence

```python
from scipy.stats import norm
```

```python
# Show (using histograms) that the binomial and normal distributions converge at large N

# Let's increase N iteratively and watch what happens to the binomial distribution:
first = np.random.binomial(20,0.5,100)
plt.hist(first)
```

```python
second = np.random.binomial(20,0.5,1000)
plt.hist(second)
```

```python
third = np.random.binomial(20,0.5,10000)
plt.hist(third)
```

```python
last = np.random.binomial(20,0.5,10000000)
plt.hist(last)
```

As N increases, the binomial distribution appears more and more normal. For comparison, here is a histogram of a normal distribution with large N:

```python
nums = np.random.normal(0,1,100000)
plt.hist(nums)
```
